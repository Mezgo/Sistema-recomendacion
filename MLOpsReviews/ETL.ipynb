```python
import pandas as pd
import numpy as np
import os
import glob
from datetime import datetime
```

```python

def getFileNames(path:str):
    files = []
    for file in os.scandir(path):
        if file.is_file():
            if file.name.endswith('.csv'):
                files.append(file.name)
                files.sort()
    return files

def joinCSV(path:str, files:list):
    df = pd.read_csv(path+files[0])
    print(df.shape)
    files.pop(0)

    for e in files:
        print(f'Using file: {e}')
        df2 = pd.read_csv(path+e)
        df = pd.concat([df,df2], ignore_index = True)
        print(df.shape)
    print(df.shape)
    return df

def joinDF(l:list):
    df = l[0]
    print(df.shape)
    l.pop(0)

    for e in l:
        df = pd.concat([df, e], ignore_index = True)
        print(df.shape)
    
    return df

def getPrefix(l:str):
    return l[0]

def idGenerator(df, prefix:str):
   print('idGenerator')
   for e in range(len(df)):
       df['id'] = prefix+df.iloc[:, 0]
   return df
        
def getDF(path:str, name:str):
    df = pd.read_csv(path+name)
    prefix = getPrefix(name)
    print(prefix)
    a = idGenerator(df, prefix)
    return a

def convert2date(date:str):
    try:
        months = {
            'january'   : '01',
            'february'  : '02',
            'march'     : '03',
            'april'     : '04',
            'may'       : '05',
            'june'      : '06',
            'july'      : '07',
            'august'    : '08',
            'septemeber': '09',
            'october'   : '10',
            'november'  : '11',
            'december'  : '12',
        }
        splt = date.split(' ')
        m = months[splt[0].lower()]
        d = splt[1]
        a = splt[2]
        date = a + '-' + m + '-' + d
        
        return date
    except:
        'null value'
```

# TRANSFORMATIONS TO THE STREAMING PLATFORMS DATA-SETS

```python
path = r'./'
streamPlatforms = getFileNames(path)
streamPlatforms
```

## Adding the new Id

```python
# en esta lista se almacenaran los df de cada plataforma
platforms = []
```

```python
# se genera el dataframe con los datos correspondientes a Amazon
Amazon = getDF(path, streamPlatforms[0])
platforms.append(Amazon) # se agrega a la lista anteriormente mencionada
```

```python
# se genera el dataframe con los datos correspondientes a Disney
Disney_plus = getDF(path, streamPlatforms[1])
platforms.append(Disney_plus)
```

```python
# se genera el dataframe con los datos correspondientes a Hulu
Hulu = getDF(path, streamPlatforms[2])
platforms.append(Hulu)
```

```python
# se genera el dataframe con los datos correspondientes a Netfix
Netflix = getDF(path, streamPlatforms[3])
platforms.append(Netflix)
```

### Joining the stream platforms data-sets

```python
# se unen los datos de las plataformas en un solo dataframe
SP_dataset = joinDF([Amazon, Disney_plus, Hulu, Netflix])
SP_dataset
```

## Filling the missing values from rating

```python
print(SP_dataset['rating'].isna().sum())
SP_dataset['rating'].replace(np.NaN, 'G', inplace=True)
print(SP_dataset['rating'].isna().sum())
```

## Formatting the date (AAAA-mm-dd)

```python
for i, e in enumerate(SP_dataset['date_added']):
    SP_dataset['date_added'].iloc[i] = convert2date(e)
```

```python
SP_dataset.info()
```

## Duration formatting

`duration_int`  -> the duration itself of the show

`duration_type` -> identify if its season or minutes

```python
SP_dataset.insert(loc = 9,
                  column ='duration_int', 
                  value = 0)
SP_dataset.insert(loc = 9, 
                  column ='duration_type', 
                  value = np.NaN)
```

```python
SP_dataset.head()
```

```python
SP_dataset['duration']
    
for i, e in enumerate(SP_dataset['duration']):
    
    if  type(e) == str:
        e = e.split(' ')
        SP_dataset['duration_type'][i] = e[1]
        SP_dataset['duration_int'][i] = int(int(e[0]))
                                                
SP_dataset
```

## Lowercasing everything

```python
f = SP_dataset.columns

for e in f:
    try:
        SP_dataset[e] = SP_dataset[e].str.lower()
    except:
        print('null value')
```
